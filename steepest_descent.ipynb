```python
from sympy import *
from sympy.plotting import plot3d

x, y = symbols('x y')
```

```python
function_1 = x**2 + y**2 - 2*x*y 
plot3d(function_1, (x, -4, 4), (y, -4, 4))
```

```python
function_2 = 2*x**2 + 2*x*y + x + y**2 - y
plot3d(function_2, (x, -4, 4), (y, -4, 4))
```

```python
def st_desc(func, x1, y1, alpha, precision, max_iter):
  printData = true

  fpx = diff(func, x)
  fpy = diff(func, y)

  grad = [fpx, fpy]
  prev_x = x1
  prev_y = y1

  iter = 0
  while True:
    new_x = prev_x - alpha*N(fpx.subs(x,prev_x).subs(y,prev_y)).evalf()
    new_y = prev_y - alpha*N(fpy.subs(y,prev_y)).subs(x,prev_x).evalf()

    iter += 1
    if iter > max_iter:
      print("Reached maximum iterations. Adjust alpha and make sure that the function is convex!")
      printData = False
      break

    # Using Convergence Criteria: 
    # 3. When the change in the design vector in two consecutive iterations is small
    if abs(new_x - prev_x) < precision and abs(new_y - prev_y) < precision:
      break

    prev_x = new_x
    prev_y = new_y

  if printData:
    print("The function "+str(func)+" converges to a minimum")
    print("Number of iterations:",iter,sep=" ")
    print("x =",new_x,sep=" ")
    print("y =",new_y,sep=" ")
```

```python
st_desc(function_1, 830, 220, .01, 1/1000000, 1000)
```

```python
st_desc(function_2, 0, 0, .01, 1/1000000, 10000)
```
